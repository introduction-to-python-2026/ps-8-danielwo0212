```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataframe = pd.read_excel("raw data.xlsx", sheet_name=0)
dataframe.to_csv("raw_data.csv", index=False)
df = dataframe[['ScreentimeHoursDayMobileTablet', 'ScreenTime_Weekly', 'OutdoorplayingHoursDay', 'Playing_week']]
df = df.dropna()

x = df['ScreentimeHoursDayMobileTablet']
y = df['OutdoorplayingHoursDay']

plt.figure(figsize=(8, 6))
plt.scatter(x, y, alpha=0.6)
plt.xlabel('ScreentimeHoursDayMobileTablet')
plt.ylabel('OutdoorplayingHoursDay')

plt.show()
```
